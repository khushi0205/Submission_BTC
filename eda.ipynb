```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
#Encoding Signal 0-none 1-buy 2-sell
signal = list(data.signal.unique())
print(signal)
for i in range(len(signal)):
    data.signal.replace({signal[i]:i},inplace=True)
```

```python
# Convert 'datetime' column to datetime data type
data['datetime'] = pd.to_datetime(data['datetime'])

# Set 'datetime' column as index
data.set_index('datetime', inplace=True)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['open'], label='Open Price', color='blue')
plt.plot(data.index, data['high'], label='High Price', color='green')
plt.plot(data.index, data['low'], label='Low Price', color='red')
plt.plot(data.index, data['close'], label='Close Price', color='black')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Bitcoin Prices Over Time')
plt.legend()

# Show plot
plt.show()
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['volume'], label='Volume', color='blue')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Bitcoin Trading Volume Over Time')
plt.legend()

# Show plot
plt.show()
```

```python
# Calculate moving averages
data['9_ema'] = data['close'].ewm(span=9).mean()
data['21_ema'] = data['close'].ewm(span=21).mean()
data['50_ema'] = data['close'].ewm(span=50).mean()
data['200_ema'] = data['close'].ewm(span=200).mean()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['close'], label='Close Price', color='blue')
plt.plot(data.index, data['9_ema'], label='9-day EMA', color='red')
plt.plot(data.index, data['21_ema'], label='21-day EMA', color='green')
plt.plot(data.index, data['50_ema'], label='50-day EMA', color='orange')
plt.plot(data.index, data['200_ema'], label='200-day EMA', color='purple')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Moving Averages of Bitcoin Prices')
plt.legend()

# Show plot
plt.show()
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Fear_and_Greed_Index'], label='Fear and Greed Index', color='blue')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Fear and Greed Index')
plt.title('Fear and Greed Index Trend Over Time')
plt.legend()

# Show plot
plt.show()
```

```python
# Plotting Relative Strength Index (RSI)
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['RSI'], label='RSI', color='blue')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('RSI')
plt.title('RSI Over Time')
plt.legend()

# Show plot
plt.show()
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['stock_to_flow_reversion'], label='Stock-to-Flow Reversion', color='blue')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Stock-to-Flow Reversion')
plt.title('Stock-to-Flow Reversion Over Time')
plt.legend()

# Show plot
plt.show()
```

```python
# Plotting spent output profit ratio
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['sth_sopr'], label='SOPR', color='blue')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('SOPR')
plt.title('SOPR Over Time')
plt.legend()

# Show plot
plt.show()
```

```python
# List of features for which histograms will be created
features = ['funding_rates', 'reserve', 'mvrv', 'nrpl', 'nupl']

# Plotting histograms for each feature
for feature in features:
    plt.figure(figsize=(8, 6))
    plt.hist(data[feature], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```
